```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (max, avg) unadjusted fitness of generation 0 = (32, 5.498)

d = []
e = []
with open('data.text', 'r') as f:
    for line in f.readlines():
        if 'mutate' in line:
            a = line.split(', ')
            # print(a)
            b = [float(s.split(': ')[1].strip()) for s in a]
            d.append(b)
        if 'max' in line:
            tup = line.split(' = ')[1].split(', ')
            max = int(tup[0][1:])
            avg = float(tup[1].strip()[:-1])
            e.append([max, avg])

df = pd.DataFrame(data = d, columns = ['mutation', 'simulation'])
ddf = pd.DataFrame(data = e[1:], columns = ['max', 'avg'])
pre_optim = pd.merge(df, ddf, right_index = True, left_index = True)


d = []
e = []
with open('data_2.text', 'r') as f:
    for line in f.readlines():
        if 'mutate' in line:
            a = line.split(', ')
            # print(a)
            b = [float(s.split(': ')[1].strip()) for s in a]
            d.append(b)
        if 'max' in line:
            tup = line.split(' = ')[1].split(', ')
            max = int(tup[0][1:])
            avg = float(tup[1].strip()[:-1])
            e.append([max, avg])

df = pd.DataFrame(data = d, columns = ['mutation', 'simulation'])
ddf = pd.DataFrame(data = e[1:], columns = ['max', 'avg'])
post_optim = pd.merge(df, ddf, right_index = True, left_index = True)
```

```python
plt.plot([i for i in range(len(pre_optim['avg']))], pre_optim['avg'], label = 'pre')
plt.plot([i for i in range(len(post_optim['avg']))], post_optim['avg'], label = 'post')
plt.legend()
plt.title('avg score of generations')
```

```python
plt.plot([i for i in range(len(pre_optim['avg']))], pre_optim['simulation'], label = 'pre')
plt.plot([i for i in range(len(post_optim['avg']))], post_optim['simulation'], label = 'post')
plt.legend()
```
